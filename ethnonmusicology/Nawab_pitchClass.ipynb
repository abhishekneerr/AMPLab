```python
import music21
from music21 import *
from music21 import analysis

import matplotlib.pyplot as plt
```

```python
in_msr = 0
out_msr = 50
```

```python
# Nawba 1, 'Btayhi_Isbahan'

s1 = converter.parse(
    'Btayhi_Isbahan.mxl')

scoreParts1 = s1.parts.stream()
for element in scoreParts1.elements:
    print(element)

part1 = scoreParts1[0]
# part0.elements
measures1 = part1.getElementsByClass(stream.Measure).stream()
print('Measures in part 0:', len(measures1))

part1_selection = part1.measures(in_msr, out_msr)
scatter1 = part1_selection.plot('scatter', 'pitchClass', 'offset')
plt.savefig(f'n1_scatter_{in_msr}-{out_msr}.png')
```
