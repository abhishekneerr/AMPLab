```python
import music21
from music21 import *
from music21 import analysis

import matplotlib.pyplot as plt
```

```python
# Nawba 1, 'Btayhi_Isbahan'

s1 = converter.parse(
    'Btayhi_Isbahan.mxl')

scoreParts1 = s1.parts.stream()
for element in scoreParts1.elements:
    print(element)

part1 = scoreParts1[0]
# part0.elements
measures1 = part1.getElementsByClass(stream.Measure).stream()
print('Measures in part 0:', len(measures1))

part1_selection = part1.measures(1, 10)
scatter1 = part1_selection.plot('scatter', 'pitchClass', 'offset')
plt.savefig('n1_scatter_1-10.png')
```
